## <b> <font color = 'yellow'>Actividad de Cierre: Implementación de Análisis Univariado </b>
<b>Objetivo:</b> Aplicar técnicas de análisis univariado para comprender la 
distribución, tendencia central y dispersión de un conjunto de datos.

<b>Conjunto de Datos: </b>Utilizaremos el conjunto de datos "Iris" de la biblioteca 
sklearn. Este dataset contiene información sobre tres especies de flores iris 
y sus características: longitud y ancho del sépalo, y longitud y ancho del 
pétalo.

## Instrucciones
### 1. Carga de Datos:
- Importa las bibliotecas necesarias: pandas, matplotlib y 
datasets de sklearn.
- Carga el conjunto de datos "Iris".

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import datasets 

# Cargar el dataset de iris
iris = datasets.load_iris() 

# Crear un DataFrame con los datos del dataset
data = pd.DataFrame(data=iris.data, columns=iris.feature_names) 
```

### <b>2. Estadísticas Descriptivas:</b>
- Selecciona una característica (por ejemplo, "longitud del 
sépalo").
- Calcula y muestra las medidas de tendencia central: media, 
mediana y moda.
- Calcula y muestra las medidas de dispersión: varianza, 
desviación estándar y rango.


```python
# Seleccionar una característica, en este caso 'longitud del sépalo'
feature = 'sepal length (cm)'

# Cálculo de estadísticas descriptivas
print("Estadísticas Descriptivas para:", feature)
print("Media:", data[feature].mean())
print("Mediana:", data[feature].median())
print("Moda:", data[feature].mode()[0])
print("Varianza:", data[feature].var())
print("Desviación Estándar:", data[feature].std())
print("Rango:", data[feature].max() - data[feature].min())
```

### <b>3. Visualización: </b>
- Crea un histograma para visualizar la distribución de la 
característica seleccionada.
- Crea un boxplot para identificar valores atípicos y comprender 
la dispersión de la característica

```python
# Histograma de la característica seleccionada
plt.hist(data[feature], bins=10, edgecolor='black')
plt.title(f'Distribución de {feature}')
plt.xlabel(feature)
plt.ylabel('Número de Flores')
plt.show()

# Boxplot para identificar valores atípicos y dispersión
plt.boxplot(data[feature])
plt.title(f'Boxplot de {feature}')
plt.ylabel(feature)
plt.show()
```

### <b>4. Interpretación: </b>
- Basándote en las estadísticas descriptivas y las visualizaciones, 
escribe un breve informe sobre la distribución de la 
característica seleccionada. ¿Hay valores atípicos? ¿Cuál es la 
tendencia central? ¿Cómo es la dispersión?

## <b> <font color = 'yellow'>Implementación de Correlación y Regresión en Análisis Multivariado</b>
<b>Objetivo:</b> Aplicar técnicas de correlación y regresión para comprender las 
relaciones entre múltiples variables en un conjunto de datos

<b>Conjunto de Datos: </b>: Utilizaremos el conjunto de datos "Iris" de la biblioteca 
sklearn. Este dataset contiene información sobre tres especies de flores iris 
y sus características: longitud y ancho del sépalo, y longitud y ancho del 
pétalo.

## Instrucciones
### 1. Carga de Datos:
- Importa las bibliotecas necesarias: pandas, matplotlib y 
datasets de sklearn.
- Carga el conjunto de datos "Iris" y conviértelo en un DataFrame.

```python
!pip install seaborn
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

# Cargar el conjunto de datos "Iris"
iris = datasets.load_iris()

# Convertir los datos a un DataFrame de pandas
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Agregar la columna de especies (target)
data['species'] = iris.target
```

### 2. Correlación entre Variables:
- Calcula la matriz de correlación entre las características del 
conjunto de datos.
- Visualiza esta matriz utilizando un heatmap.

```python
# Calcular la matriz de correlación entre las características
correlation_matrix = data.corr()

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Título del gráfico
plt.title('Matriz de Correlación')

# Mostrar el gráfico
plt.show()
```

### 1. Carga de Datos
Primero, importamos las bibliotecas necesarias y cargamos el conjunto de datos Iris.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from scipy.stats import kurtosis, skew
import statsmodels.api as sm

# Cargar el conjunto de datos Iris
iris = datasets.load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

### 2. Selección de la Característica
Seleccionamos "longitud del sépalo" para nuestro análisis.

```python
feature = 'sepal length (cm)'
```

### 3. Estadísticas Descriptivas
Calculamos las medidas de tendencia central y de dispersión.

```python
# Tendencia central
media = data[feature].mean()
mediana = data[feature].median()
moda = data[feature].mode()[0]

# Medidas de dispersión
varianza = data[feature].var()
desviacion_estandar = data[feature].std()
rango = data[feature].max() - data[feature].min()
skewness = skew(data[feature])
kurt = kurtosis(data[feature])

# Imprimir resultados
print("Tendencia Central:")
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

print("\nMedidas de Dispersión:")
print(f"Varianza: {varianza}")
print(f"Desviación Estándar: {desviacion_estandar}")
print(f"Rango: {rango}")
print(f"Asimetría (Skewness): {skewness}")
print(f"Curtosis: {kurt}")
```

### 4. Visualizaciones
#### 4.1 Histograma
Creamos un histograma para visualizar la distribución de "longitud del sépalo".

```python
plt.figure(figsize=(10, 5))
plt.hist(data[feature], bins=10, edgecolor='black', alpha=0.7)
plt.title(f'Distribución de {feature}')
plt.xlabel(feature)
plt.ylabel('Número de Flores')
plt.show()
```

#### 4.2 Boxplot
Creamos un boxplot para identificar valores atípicos y comprender la dispersión.

```python
plt.figure(figsize=(10, 5))
plt.boxplot(data[feature])
plt.title(f'Boxplot de {feature}')
plt.ylabel(feature)
plt.show()
```

#### 4.3 Diagrama de Violín
Usamos un diagrama de violín para mostrar la distribución y los cuartiles.

```python
plt.figure(figsize=(10, 5))
sns.violinplot(y=data[feature])
plt.title(f'Diagrama de Violín de {feature}')
plt.ylabel(feature)
plt.show()
```

#### 4.4 Gráfico de Densidad
Visualizamos la distribución con un gráfico de densidad.

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data[feature])
plt.title(f"Gráfico de Densidad de {feature}")
plt.show()
```

#### 4.5 QQ Plot
Comprobamos la normalidad de la variable usando un QQ Plot.

```python
plt.figure(figsize=(10, 5))
sm.qqplot(data[feature], line='45')
plt.title(f"QQ Plot de {feature}")
plt.show()
```

### 5. Interpretación
- Distribución: El histograma y el gráfico de densidad sugieren que la longitud del sépalo sigue una distribución aproximadamente normal, aunque con una ligera asimetría positiva.
- Valores Atípicos: El boxplot muestra algunos valores atípicos, pero no son significativos en comparación con la mayoría de los datos.
- Tendencia Central: La media y la mediana son bastante cercanas, lo que sugiere que la tendencia central no se ve afectada por los valores atípicos.
- Dispersión: La desviación estándar indica que hay cierta variabilidad en la longitud del sépalo, pero el rango muestra que la longitud de los sépalos varía de manera relativamente controlada.
