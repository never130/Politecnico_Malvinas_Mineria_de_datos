# Tablas de Distribución de Frecuencias

Las tablas de distribución de frecuencias son herramientas estadísticas que organizan y presentan datos para facilitar la visualización de la distribución de valores en un conjunto de datos. Veamos más sobre su uso, consejos para su armado y cómo calcular los intervalos.

## Uso de las Tablas de Distribución de Frecuencias

- **Análisis de Datos**: Facilitan la comprensión de la distribución de los datos, mostrando dónde se concentran los valores, las tendencias y posibles anomalías.
- **Preparación para Visualizaciones**: Son útiles para crear gráficos como histogramas, que muestran las frecuencias de los intervalos.
- **Resumen de Grandes Conjuntos de Datos**: Permiten condensar grandes volúmenes de datos en un formato más manejable y comprensible.

## Consejos para el Armado de Tablas de Distribución de Frecuencias

1. **Claridad**: Mantén la tabla simple y fácil de leer, con títulos claros para cada columna.
2. **Orden**: Ordena las categorías de forma lógica si los datos son nominales o ordinales.
3. **Consistencia**: Procura utilizar intervalos de igual tamaño para facilitar la comparación entre ellos.

## Cómo Calcular los Intervalos

El cálculo de los intervalos es crucial y puede influir significativamente en la interpretación de los datos. Aquí te muestro cómo hacerlo:

### Regla de Sturges
Se utiliza la fórmula \( k = 1 + 3.322 \log_{10} n \), donde \( k \) es el número de intervalos y \( n \) el número total de observaciones. Esta fórmula es útil para conjuntos de datos de tamaño moderado.

### Amplitud del Intervalo
Una vez que se determina el número de intervalos, la amplitud se puede calcular como:
\[ \text{Amplitud} = \frac{\text{Valor Máximo} - \text{Valor Mínimo}}{k} \]

### Ejemplo de Cálculo de Intervalos
Si tienes datos con un rango de edades de 10 a 70 años y 100 observaciones:

1. Calcula \( k \) con la Regla de Sturges: \( k = 1 + 3.322 \log_{10} 100 \approx 8 \)
2. Calcula la amplitud: \[ \text{Amplitud} = \frac{70 - 10}{8} = 7.5 \]

Estos pasos ayudarán a distribuir los datos en 8 intervalos, cada uno con un rango de 7.5 años, simplificando la interpretación y ayudando a identificar tendencias en la distribución de las edades.


# Creación de un Histograma con Python

En este notebook, aprenderemos cómo crear un histograma utilizando las bibliotecas Pandas, Seaborn y Matplotlib en Python. Un histograma es útil para visualizar la distribución de datos numéricos mediante el agrupamiento de datos en 'bins' o intervalos.

## Importar Bibliotecas

Comenzaremos importando las bibliotecas necesarias para nuestro análisis y visualización.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Crear Datos de Ejemplo

Para este ejemplo, generaremos un conjunto de datos aleatorios que representen las edades de un grupo de personas.



```python
# Generar datos aleatorios
import numpy as np
np.random.seed(25)
edades = np.random.normal(45, 20, 200)  # 200 edades aleatorias entre 0 y 100

# Crear un DataFrame de Pandas
df = pd.DataFrame(edades, columns=['Edades'])
df.describe()
```

## Crear un Histograma con Pandas

Usaremos Pandas para crear un histograma directamente desde el DataFrame.


```python
# Crear un histograma con Pandas
df['Edades'].hist(bins='auto', color='blue', edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.grid(False)
plt.show()
```

## Crear un Histograma con Seaborn

Seaborn es otra biblioteca de visualización de datos que ofrece una interfaz de alto nivel para dibujar gráficos estadísticos atractivos.


```python
# Crear un histograma con Seaborn
sns.histplot(data=df, x='Edades', bins='auto', color='red', kde=True)
plt.title('Distribución de Edades con KDE')
plt.xlabel('Edades')
plt.ylabel('Densidad')
plt.show()
```

 # Boxplot o Diagrama de Caja y Bigotes

El boxplot, también conocido como diagrama de caja y bigotes, es una representación gráfica que muestra la distribución de los datos numéricos a través de cuartiles. Este diagrama es útil para resumir una gran cantidad de datos numéricos mediante la visualización de su mediana, cuartiles, valores mínimos y máximos, y valores atípicos (outliers).

## Utilidades del Boxplot

- **Resumen Visual de Datos**: El boxplot sintetiza información clave de la distribución de los datos, como la mediana, los cuartiles y los extremos, facilitando la comprensión rápida del conjunto de datos.
- **Identificación de Outliers**: Los valores que no se incluyen entre los bigotes se consideran outliers y son visualmente destacados como puntos individuales, lo cual es crucial para la depuración de datos o análisis detallados.
- **Comparación de Distribuciones**: Es particularmente útil para comparar las distribuciones entre diferentes grupos o condiciones en un conjunto de datos, mostrando cómo varían las medias y la dispersión.
- **Visualización de Dispersión y Asimetría**: El boxplot muestra la dispersión de los datos, la mediana y proporciona indicios sobre la simetría de la distribución de los datos.

## Componentes de un Boxplot

- **Mediana (Q2/50th Percentil)**: La línea dentro del cajón indica la mediana de la distribución.
- **Cuartiles (Q1 y Q3)**: El borde inferior y superior del cajón representan el primer (25%) y tercer cuartil (75%) de los datos, respectivamente.
- **Bigotes**: Las líneas que se extienden desde el cajón hasta los valores máximo y mínimo no atípicos.
- **Outliers**: Representados generalmente por puntos fuera de los bigotes, indican valores que difieren significativamente del resto de los datos.

Este gráfico es una herramienta poderosa no solo para análisis estadístico básico, sino también para observaciones detalladas en estudios comparativos y exploratorios.


## Generar Datos de Ejemplo

Crearemos algunos datos simulados que representen, por ejemplo, las puntuaciones de diferentes grupos en una prueba.


```python
# Generar datos aleatorios
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

a = data1.tolist()
print(type(a))

# Crear un DataFrame
df = pd.DataFrame({'Grupo 1': data1, 'Grupo 2': data2, 'Grupo 3': data3})
```

## Crear un Boxplot con Matplotlib

Seaborn facilita la creación de boxplots y añade automáticamente más estilo y formato.

# Scatterplots o Diagramas de Dispersión

Los scatterplots, o diagramas de dispersión, son representaciones gráficas que utilizan puntos para mostrar los valores de dos variables diferentes para un conjunto de datos. La posición de cada punto en el eje horizontal y vertical indica los valores de las dos variables. Este tipo de gráfico es fundamental para explorar relaciones o correlaciones entre variables numéricas.

## Utilidades del Scatterplot

- **Exploración de Relaciones**: Los scatterplots son ideales para identificar visualmente las relaciones o correlaciones entre dos variables. Si los puntos forman un patrón, esto sugiere una relación.
- **Identificación de Comportamientos Anómalos**: Además de las tendencias generales, los scatterplots pueden ayudar a identificar valores atípicos o comportamientos anómalos en los datos.
- **Comparación de Grupos**: Al colorear los puntos según una categoría, los scatterplots pueden mostrar cómo se distribuyen los grupos de datos en relación con las variables estudiadas.
- **Modelado de Regresión**: Son útiles para visualizar el ajuste de modelos de regresión, mostrando la distancia de los puntos a la línea de mejor ajuste.

## Componentes de un Scatterplot

- **Ejes**: Cada eje representa una de las variables. La elección de qué variable colocar en cada eje depende del análisis que se está realizando.
- **Puntos**: Cada punto en el gráfico representa una observación en el conjunto de datos.
- **Líneas de Tendencia**: A veces se añaden al gráfico para indicar la relación general entre las variables.
- **Color/Estilo de Puntos**: Se pueden usar diferentes colores o estilos para representar diferentes categorías o grupos dentro de los datos, proporcionando otra dimensión de análisis.

Estos diagramas son especialmente valiosos en la estadística para verificar supuestos, analizar patrones y sugerir hipótesis sobre la causalidad.

## Ejemplo Básico de un Scatterplot

Supongamos que queremos visualizar la relación entre el tiempo de estudio y las calificaciones obtenidas por los estudiantes.


```python
import matplotlib.pyplot as plt

# Datos de ejemplo
tiempo_estudio = [1, 2, 4, 5, 7, 8, 10]  # Horas de estudio
calificaciones = [55, 60, 65, 70, 80, 85, 90]  # Calificaciones obtenidas

# Crear scatterplot
plt.figure(figsize=(8, 5))
plt.scatter(tiempo_estudio, calificaciones, color='blue')
plt.title('Relación entre Tiempo de Estudio y Calificaciones')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```

```python
# Datos de ejemplo
data = {
    'Horas de Estudio': [1, 2, 4, 5, 7, 8, 10],
    'Calificaciones': [55, 60, 65, 70, 80, 85, 90],
    'Género': ['F', 'M', 'F', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)

# Crear scatterplot con Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horas de Estudio', y='Calificaciones', hue='Género', style='Género', palette='deep', s=100)
plt.title('Relación entre Horas de Estudio y Calificaciones por Género')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.legend(title='Género')
plt.show()
```

# Introducción al QQ Plot (Quantile-Quantile Plot)

El QQ Plot, o Quantile-Quantile Plot, es una herramienta gráfica utilizada para comparar dos distribuciones probabilísticas trazando sus cuantiles respectivos uno contra otro. Este tipo de gráfico es fundamental en estadísticas para evaluar si un conjunto de datos se ajusta a una distribución teórica esperada, como la normal, o para comparar distribuciones entre dos conjuntos de datos.

## Utilidad del QQ Plot

- **Verificación de Normalidad**: Uno de los usos más comunes del QQ Plot es verificar la hipótesis de normalidad de los datos. Esto es crucial en métodos estadísticos que asumen normalidad de los datos, como muchas pruebas paramétricas y análisis de regresión.
- **Comparación de Distribuciones**: Permite comparar de manera visual la distribución de un conjunto de datos con una distribución teórica o con otro conjunto de datos, lo cual es útil para identificar si dos muestras provienen de la misma población o distribución.
- **Diagnóstico en Análisis de Regresión**: En análisis de regresión, los QQ Plots son utilizados para evaluar los supuestos de normalidad en los residuos de un modelo.

## ¿Cuándo usar un QQ Plot?

Un QQ Plot es particularmente útil cuando necesitas:
- **Evaluar la Normalidad de los Datos**: Antes de aplicar pruebas estadísticas que requieren normalidad.
- **Comparar Distribuciones Teóricas y Empíricas**: Para confirmar supuestos teóricos sobre la distribución de tus datos.
- **Investigar la Homogeneidad de Varianzas**: En análisis comparativos, para verificar si diferentes muestras podrían haber venido de poblaciones con la misma varianza.

El QQ Plot es una herramienta esencial en el kit de herramientas de cualquier analista de datos, proporcionando una base sólida para la toma de decisiones y análisis estadísticos avanzados.


### Ejemplo de QQ Plot con Matplotlib
Matplotlib no tiene una función incorporada específicamente para QQ plots, pero podemos utilizar el módulo stats de SciPy para calcular los cuantiles y luego trazarlos.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generar datos de ejemplo
data = np.random.exponential(size=1000)

# Generar QQ plot
fig, ax = plt.subplots()
res = stats.probplot(data, dist="norm", plot=ax)
ax.get_lines()[1].set_color('red')
ax.set_title('QQ Plot de Datos Normales')
ax.set_xlabel('Cuantiles Teóricos')
ax.set_ylabel('Cuantiles de la Muestra')
plt.show()
```

### QQ Plot en Seaborn
Seaborn no ofrece una función específica para QQ plots, y por lo general se utiliza Matplotlib o statsmodels para este propósito. Sin embargo, se puede hacer un aproximado utilizando técnicas de visualización scatter si se calculan los cuantiles manualmente.

```python
import seaborn as sns

# Calcular cuantiles
quantiles = np.percentile(data, np.linspace(0, 100, num=100))

norm_quantiles = np.percentile(np.random.normal(loc=0, scale=1, size=1000), np.linspace(0, 100, num=100))

# Crear scatterplot de cuantiles
plt.figure(figsize=(8, 6))
sns.scatterplot(x=norm_quantiles, y=quantiles)
plt.title('Aproximación de QQ Plot con Seaborn')
plt.xlabel('Cuantiles Normales Teóricos')
plt.ylabel('Cuantiles de la Muestra')
plt.plot(norm_quantiles, norm_quantiles, color="red", linestyle="--") # Línea teórica
plt.show()
```

# Introducción a Otros Gráficos Estadísticos Comunes

Además de los gráficos más conocidos como histogramas, scatterplots y QQ plots, existen otros tipos de visualizaciones que pueden ofrecer insights valiosos en diferentes contextos de análisis de datos. A continuación, exploraremos algunos de estos gráficos, sus usos y consejos para su implementación eficaz.

## Otros Gráficos y Sus Usos

1. **Gráficos de Barras y Barras Apiladas**: Útiles para comparar cantidades en diferentes categorías y visualizar la composición de cada categoría.
2. **Gráficos de Líneas**: Ideales para visualizar tendencias a lo largo del tiempo, especialmente útiles en análisis de series temporales.
3. **Gráficos de Área**: Similares a los gráficos de líneas, pero con el área bajo la línea rellena, enfatizando la magnitud de los cambios a lo largo del tiempo.
4. **Heatmaps (Mapas de Calor)**: Perfectos para mostrar la magnitud de un fenómeno en dos dimensiones mediante el uso de colores, comúnmente utilizados en matrices de correlación o análisis de datos categóricos complejos.

## Consejos para la Creación de Gráficos

- **Claridad es Clave**: Los gráficos deben transmitir su mensaje de forma clara y directa, sin ambigüedades.
- **Etiquetado Adecuado**: Incluye títulos, etiquetas en los ejes y leyendas claras y precisas para facilitar la comprensión de los datos presentados.
- **Color con Propósito**: Utiliza el color no solo para hacer el gráfico atractivo, sino para mejorar la comprensión y el análisis de los datos.
- **Evitar Sobrecargar**: Un gráfico sobrecargado de datos o categorías puede resultar confuso y difícil de interpretar. Simplifica cuando sea necesario.

Estos consejos y ejemplos te ayudarán a elegir y diseñar los gráficos más apropiados para tus necesidades de análisis de datos, asegurando que la información presentada sea accesible y útil.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos de ejemplo para gráficos de barras
categorias = ['Categoría 1', 'Categoría 2', 'Categoría 3']
valores = [10, 15, 8]

plt.figure(figsize=(10, 6))
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras Simple')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
# Datos de ejemplo para gráfico de líneas
fechas = pd.date_range(start='1/1/2020', periods=365)
datos = np.random.rand(365).cumsum()

plt.figure(figsize=(10, 6))
plt.plot(fechas, datos, marker='o', color='purple', linewidth=2)
plt.title('Gráfico de Líneas para Datos Temporales')
plt.xlabel('Fecha')
plt.ylabel('Valor Acumulado')
plt.grid(True)
plt.show()
```

```python
# Datos para heatmap
data = np.random.rand(10, 12)
sns.heatmap(data, cmap='coolwarm')
plt.title('Heatmap de Datos Aleatorios')
plt.show()
```
