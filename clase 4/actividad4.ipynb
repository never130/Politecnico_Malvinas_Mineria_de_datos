## <font color='yellow'> Visualización de Datos en Minería de Datos </font>### Grafico de QQPlot y BoxPlot1. En la primera celda de tu cuaderno, escribe y ejecuta el siguiente 
código para importar las bibliotecas necesarias:



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

2. Utilizaremos el conjunto de datos 'Iris', ampliamente utilizado en la 
comunidad de minería de datos. Puedes cargarlo directamente desde 
la biblioteca Seaborn:


```python
iris = sns.load_dataset('iris')
```

3. QQ-Plot: El QQ-Plot (Quantile-Quantile Plot) se utiliza para 
comparar la forma de una distribución con una distribución normal. 
Es útil para verificar la normalidad de los datos.
Genera un QQ-Plot para la variable sepal_length:

```python
stats.probplot(iris['sepal_length'], dist="norm", plot=plt)
plt.title('QQ-Plot para la longitud del sépalo')
plt.show()
```

4. Boxplot: El Boxplot proporciona un resumen gráfico de la 
distribución de una variable, mostrando la mediana, los cuartiles y los 
valores atípicos.


```python
# Creamos el boxplot para la variable 'sepal_length'
sns.boxplot(x=iris['sepal_length'])
plt.title('Boxplot para la longitud del sépalo')
plt.xlabel('Longitud del sépalo (cm)')
plt.show()
```

### Interpretacion:
En un boxplot, los valores atípicos son aquellos que están por debajo del cuartil inferior (Q1) menos 1.5 veces el rango intercuartílico (IQR), o por encima del cuartil superior (Q3) más 1.5 veces el IQR. Si no hay puntos fuera de los bigotes, entonces no hay valores atípicos evidentes para la variable sepal_length.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Cargar el conjunto de datos Titanic de Seaborn
titanic = sns.load_dataset('titanic')

# Mostrar las primeras filas del DataFrame para verificar la estructura
print(titanic.head())
```

```python
# Crear un scatter plot para visualizar la relación entre la tarifa pagada y la edad según supervivencia
plt.figure(figsize=(12, 8))
colors = {0: 'red', 1: 'blue'}

# Crear el scatter plot con la personalización de colores
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic, palette=colors)

# Modificar la leyenda
handles, labels = plt.gca().get_legend_handles_labels()
labels = ['No sobrevivió', 'Sobrevivió']  # Cambiar las etiquetas
plt.legend(handles, labels, title='Supervivencia', loc='upper right')

plt.title('Relación entre Edad y Tarifa Pagada según la Supervivencia')
plt.xlabel('Edad')
plt.ylabel('Tarifa Pagada')
plt.show()
```

```python
# Crear un QQ plot para verificar la normalidad en la distribución de la edad
plt.figure(figsize=(8, 8))
stats.probplot(titanic['age'].dropna(), dist="norm", plot=plt)
plt.title('QQ Plot para la Distribución de la Edad')
plt.xlabel('Cuantiles Teóricos')
plt.ylabel('Cuantiles de la Muestra')
plt.show()
```

El scatter plot muestra la relación entre la edad de los pasajeros y la tarifa que pagaron por su billete. Los puntos en azul representan a aquellos que sobrevivieron, mientras que los puntos en rojo indican a los que no sobrevivieron. Este gráfico revela patrones interesantes, como que muchos pasajeros jóvenes que pagaron tarifas más altas tienden a haber sobrevivido, mientras que algunos pasajeros mayores pagaron tarifas similares y no sobrevivieron. Este tipo de visualización nos ayuda a comprender mejor los factores que influyeron en la supervivencia durante el hundimiento del Titanic.



Reflexión sobre las Técnicas de Visualización
1. Scatter Plot (Diagrama de Dispersión)
El diagrama de dispersión que relaciona la edad y la tarifa pagada según la supervivencia me permitió identificar patrones en cómo la edad y el costo del pasaje influyen en la probabilidad de supervivencia. Al ver la distribución de los puntos, noté que:

Los pasajeros más jóvenes que pagaron tarifas más altas tienden a sobrevivir más, lo que puede sugerir que aquellos con más recursos económicos (que suelen ser más jóvenes) tuvieron mejores oportunidades de sobrevivir.
También se observaron grupos de pasajeros mayores que pagaron tarifas más bajas, lo que indica que estos podrían haber enfrentado más desafíos durante el naufragio.
Este tipo de visualización hace que sea fácil ver relaciones y tendencias que pueden no ser evidentes en los datos tabulares.

2. QQ Plot
El QQ plot utilizado para verificar la normalidad en la distribución de la edad me permitió evaluar si los datos seguían una distribución normal, que es una suposición común en muchas pruebas estadísticas.

Al observar el gráfico, pude ver si los puntos se alinean a lo largo de la línea diagonal, lo que indicaría que la distribución de la edad es normal. Si los puntos se desvían significativamente de la línea, eso sugiere que los datos no son normales, lo que puede influir en la selección de análisis estadísticos a utilizar.
